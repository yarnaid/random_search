```python
perim = 100
```

```python
from optimisation import climbing, annealing, genetic
```

```python
import numpy as np
```

```python
init_function = lambda : abs(np.random.random() * perim)
```

```python
move = lambda x: [abs(x+np.random.random()), abs(x-np.random.random())]
```

```python
move(10)
```

```python
objective_function = lambda x: x*(perim/2.-x)
```

```python
num_evaluations, best_score, best = climbing.hillclimb_and_restart(
    init_function,
    move,
    objective_function,
    max_evaluations=100
)
print num_evaluations, best_score, best
```

```python
class BestSide(annealing.Annealer):
    def __init__(self):
        super(BestSide, self).__init__(init_function())
        
    def energy(self):
        return -objective_function(self.state)  # we minimize

    def move(self):
        self.state = abs(self.state + (np.random.random() - 0.5)*2)
bs = BestSide()
bs.steps = 200
state, area = bs.anneal()
print "Sides: ", state, "x", perim/2.-state, "\nArea: ", -area
```

```python
def get_random_genome():
    x = np.random.random()*perim
    return [x, perim/2-x]
```

```python
def fitness(x):
    res = x[0] * x[1]
    if x[0] + x[1] > perim/2:
        res = 0
    return res
```

```python
parents_num = 8
parents = [get_random_genome() for x in xrange(parents_num)]
best_x, best_score = genetic.run(parents, fitness, ([1, perim/2.], [1, perim/2.]))
print "Best x:", best_x, "\nBest area: ", best_score
```
