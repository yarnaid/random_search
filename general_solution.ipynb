# General Solution for 3 Methods

```python
from pybrain.optimization import GA, StochasticHillClimber, HillClimber
from pybrain.tools.shortcuts import buildNetwork
from pybrain.structure import SigmoidLayer, TanhLayer, SoftmaxLayer
```

```python
def solve(optimizer, dataset, test_dataset=None, hidden_number=1, hidden=SoftmaxLayer, minimize=True, steps=20, verbose=True, temperature=1.0):

    neurons_number = int(dataset.indim)
    indim = dataset.indim
    outdim = dataset.outdim
    
    nn = buildNetwork(indim, neurons_number, neurons_number, outdim, hiddenclass=hidden)

    if test_dataset is None:
        test_data, train_data = dataset.splitWithProportion(0.25)
    
        try:
            test_data._convertToOneOfMany()  # This is necessary, but I don't know why
            train_data._convertToOneOfMany()  # http://stackoverflow.com/q/8154674/562769
        except:
            pass
    else:
        train_data = dataset
        test_data = test_dataset

    
    opt = optimizer(train_data.evaluateModuleMSE, nn, minimize=minimize, verbose=verbose, temperature=temperature)
    
    nn, score = opt.learn(steps)

    for inp, trg in test_data:
        print nn.activate(inp)[0], trg
    
    return nn
```

# Salary Problem

```python
from pybrain.datasets import SupervisedDataSet
```

```python
import numpy as np
import pandas as pd
features_df = pd.read_csv('./data/train_features_5000.csv')
salary_df = pd.read_csv('./data/train_salaries_5000.csv')

df = features_df.merge(salary_df, on='jobId')
```

```python
def col_to_int(col_name, df):
    gr = df.groupby(df[col_name])
    values_dict = dict()
    for k in gr.groups:
        values_dict[k] = df.ix[gr.groups[k]].salary.mean()
    col = df[col_name]
    convert = lambda x: values_dict[x]
    return col.apply(convert)

def col_to_int_random(col_name, df):
    col = df[col_name]
    values = set(col)
    numbers = np.arange(len(values))
    values_dict = dict(zip(values, numbers))
    convert = lambda x: values_dict[x]
    return col.apply(convert)

cols_to_int = set(df.columns.values[1:-1])
for col in cols_to_int:
    df[col] = col_to_int(col, df)
cols_to_int = set(df.columns.values[:1])
for col in cols_to_int:
    df[col] = col_to_int_random(col, df)
```

```python
x = df[df.columns[:-1]].copy().as_matrix()
y = salary_df.salary.copy().as_matrix()
```

```python
x_dim = len(x[0])
out_dim = 1

ds = SupervisedDataSet(x_dim, out_dim)
```

```python
x[0]
```

```python
for i in xrange(len(x)):
    xx = x[i]
    yy = (y[i], )
    ds.addSample(xx, yy)
```

```python
solve(GA, ds, steps=5)
```

```python
solve(HillClimber, ds, steps=20)
```

```python
solve(StochasticHillClimber, ds, steps=50, temperature=1000)
```

# Cancer Problem

```python
from pybrain.datasets import ClassificationDataSet
```

```python
import numpy as np
dataset = np.loadtxt('./data/breast-cancer-train.csv', skiprows=1, delimiter=',')
x = dataset[:, :-1]
y = dataset[:, -1]

x_dim = len(x[0])
out_dim = 1

dataset_test = np.loadtxt('./data/breast-cancer-test.csv', skiprows=1, delimiter=',')
xt = dataset_test[:, :-1]
yt = dataset_test[:, -1]


ds = ClassificationDataSet(x_dim, out_dim, nb_classes=2)
ds_test = ClassificationDataSet(x_dim, out_dim, nb_classes=2)
```

```python
for i in xrange(len(x)):
    xx = x[i]
    yy = [y[i]]
    ds.addSample(xx, yy)
```

```python
for i in xrange(len(xt)):
    xx = xt[i]
    yy = [yt[i]]
    ds_test.addSample(xx, yy)
```

```python
solve(GA, ds, test_dataset=ds_test, steps=20)
```

```python
solve(HillClimber, ds, test_dataset=ds_test, steps=50)
```
