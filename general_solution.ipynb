# General Solution for 3 Methods

```python
from pybrain.optimization import GA, StochasticHillClimber, HillClimber
from pybrain.tools.shortcuts import buildNetwork
from pybrain.structure import SigmoidLayer, TanhLayer, SoftmaxLayer
```

```python
class Listener(object):
    '''
    saving resutls
    '''
    def __init__(self):
        self.scores = []
        self.evaluable = []
    def __call__(self, evaluable, scores):
        self.scores.append(scores)
        self.evaluable.append(evaluable)
```

```python
def solve(optimizer, dataset, test_dataset=None, hidden_number=1, hidden=SoftmaxLayer, minimize=True, steps=20, listener=None, verbose=True, temperature=1.0):
    '''
    General solver for all problems. It takes optimizer with params and give him dataset and neural network.
    Than it trains and get results
    '''
    neurons_number = int(dataset.indim)
    indim = dataset.indim
    outdim = dataset.outdim
    
    nn = buildNetwork(indim, neurons_number, neurons_number, outdim, hiddenclass=hidden)

    if test_dataset is None:
        data_length = len(dataset)
        data_proportion = min(20/data_length, 0.25)
        test_data, train_data = dataset.splitWithProportion(0.25)
    
        try:
            test_data._convertToOneOfMany()  # This is necessary, but I don't know why
            train_data._convertToOneOfMany()  # http://stackoverflow.com/q/8154674/562769
        except:
            pass
    else:
        train_data = dataset
        test_data = test_dataset

    
    opt = optimizer(train_data.evaluateModuleMSE, nn, minimize=minimize, verbose=verbose, temperature=temperature,
                    listener=listener)
    
    nn, score = opt.learn(steps)

    if verbose:
        for inp, trg in test_data:
            print nn.activate(inp)[0], trg
    
    return nn
```

```python
from time import time

def experiment(number, **params):
    ''' returns results of experiments for statistic '''
    scores = list()
    evs = list()
    times = []
    start_total = time()
    for i in xrange(number):
        listener = Listener()
        params.update({'listener': listener})
        start_time = time()
        solve(**params)
        worked = time() - start_time
        scores.append(listener.scores)
        evs.append(listener.evaluable)
        times.append(worked)
    print "Finished in {} seconds".format(time() - start_total)
    return {'scores': scores, 'evs': evs, 'time': times}
        
    
```

# Salary Problem

### Load and Clean Data

```python
from pybrain.datasets import SupervisedDataSet
```

```python
import numpy as np
import pandas as pd
features_df = pd.read_csv('./data/train_features_5000.csv')
salary_df = pd.read_csv('./data/train_salaries_5000.csv')

df = features_df.merge(salary_df, on='jobId')
```

```python
def col_to_int(col_name, df):
    gr = df.groupby(df[col_name])
    values_dict = dict()
    for k in gr.groups:
        values_dict[k] = df.ix[gr.groups[k]].salary.mean()
    col = df[col_name]
    convert = lambda x: values_dict[x]
    return col.apply(convert)

def col_to_int_random(col_name, df):
    col = df[col_name]
    values = set(col)
    numbers = np.arange(len(values))
    values_dict = dict(zip(values, numbers))
    convert = lambda x: values_dict[x]
    return col.apply(convert)

cols_to_int = set(df.columns.values[1:-1])
for col in cols_to_int:
    df[col] = col_to_int(col, df)
cols_to_int = set(df.columns.values[:1])
for col in cols_to_int:
    df[col] = col_to_int_random(col, df)
```

```python
x = df[df.columns[:-1]].copy().as_matrix()
y = salary_df.salary.copy().as_matrix()
```

```python
x_dim = len(x[0])
out_dim = 1

ds = SupervisedDataSet(x_dim, out_dim)
```

```python
print "Run HC"
hc_res = experiment(exp_number, optimizer=HillClimber, dataset=ds, steps=steps, verbose=False)
print "Run An"
an_res = experiment(exp_number, optimizer=StochasticHillClimber, dataset=ds, steps=steps, temperature=1000, verbose=False)
print "Run GA"
ga_res = experiment(exp_number, optimizer=GA, dataset=ds, steps=steps, verbose=False)
```
