# Salesman Problem

The problem as following (as you probably know).
There is a map of some area with cities, which salesman want to visit.
And salesman is lazy and he want to have the shortest way and at the end return to home.

Let's generate random map and will create tours of salesman.

```python
from matplotlib.path import Path
import matplotlib.patches as patches
from matplotlib import  pyplot as plt
import seaborn as sns
sns.set(style="white")
%matplotlib inline
```

```python
import numpy as np

def create_map(max_x, max_y, cities_num):
    '''create map of the cities with center of map in mean_x, mean_x'''
    res = np.random.random((cities_num, 2))
    res.T[0] = res.T[0] * max_x
    res.T[1] = res.T[1] * max_y
    return res.T
    # for gaussian cities. They normal too, but too complicated for viewing. You can try them, if you want
    rs = np.random.RandomState(9)
    mean = [max_x/2., max_y/2.]
    cov = [(1, 0), (0, 1)]
    res = rs.multivariate_normal(mean, cov, cities_num).T  #just math, no magic    
    return res # [[x1,x2,x3....], [y1,y2,y3,...]]

max_x = 10000
max_y = 10000
cities_num = 20
mp = create_map(max_x, max_y, cities_num)
points = mp.T # [[x1,y1],[x2,y2],...]
x, y = mp

color = sns.color_palette()[1]
grid = sns.JointGrid(x, y, space=0, size=6, ratio=50)
grid.plot_joint(plt.scatter, color=color, alpha=.8)
grid.plot_marginals(sns.rugplot, color=color)
```

On the next step we create distance matrix

```python
def get_distance(p1, p2):
    return np.sqrt(sum((p2-p1)**2))
print "Test distance: ", get_distance(points[0], points[1])

def get_distance_matrix(points, get_distance=get_distance):
    '''
    creates matrix of distances between i and j points in input array
    '''
    points_num = len(points)
    res = np.zeros((points_num, points_num))
    for i in xrange(points_num):
        p_i = points[i]
        for j in xrange(points_num):
            p_j = points[j]
            res[i][j] = get_distance(p_i, p_j)
    return res

dist_mat = get_distance_matrix(points)
```

```python
def get_init_tour(points):
    '''
    create random tour
    '''
    tour = np.arange(len(points))
    np.random.shuffle(tour)
    return tour
```

```python
def get_tour_length(tour, dist_mat=dist_mat):
    '''
    fitness/objective function to minimize
    '''
    res = 0
    try:
        for i in xrange(0, len(tour)-1):
            res += dist_mat[tour[i]][tour[i+1]]  # distaces between points i and i+1 in tour
        res += dist_mat[tour[-1]][tour[0]]  # to return home
    except:
        res = dist_mat.max() * len(tour)
    if len(set(tour)) != len(tour):
        res = dist_mat.max() * len(tour)
    return res
print get_tour_length([0,1]), get_distance(points[0], points[1]) + get_distance(points[1], points[0])
```

So in such problem domain we have $\pmb{x}$ is the tour and $f(\pmb{x})$ is `get_tour_length` function.

To use it with `pybran` algorithms we have to implement new class as subclass of `ParameterContainer`.

```python
from pybrain.structure.parametercontainer import ParameterContainer

class SalesMan(ParameterContainer):
    def __init__(self, x):
        super(SalesMan, self).__init__(len(x), stdParams=0)
        self._params = x  # tour
    def mutate(self):
        '''swap 2 points in tour'''
        max_pos = len(self._params)
        pos1 = np.random.randint(1, max_pos-1)  # to fix home
        pos2 = pos1 + 1
        self._params[pos1], self._params[pos2] = self._params[pos2], self._params[pos1]
    def randomize(self):
        '''shuffle tour'''
        res = np.arange(len(self._params))
        np.random.shuffle(res)
        self._params = res
    def copy(self):
        return SalesMan(self._params)
    
    def resetDerivatives(self):
        '''
        not needed, not to be this class abstract
        '''
        assert self.hasDerivatives
        deriv = np.arange(len(self._params))
        np.random.shuffle(deriv)
        self._derivs = deriv
    def __repr__(self):
        return 'tour length: {}, tour: {}'.format(str(get_tour_length(self._params)), self._params)
```

In `pybrain` GA is for continuous problems. We should redefine it.

```python
from pybrain.optimization import GA, StochasticHillClimber, HillClimber
```

```python
from random import choice, random

class DGA(GA):
    def initPopulation(self):
        if self.initialPopulation is not None:
            self.currentpop = self.initialPopulation
        else:
            self.currentpop = [self._initEvaluable]
            for _ in range(self.populationSize-1):
                home = self._initEvaluable[0]
                new_indiv = np.arange(0, len(self._initEvaluable))
                np.random.shuffle(new_indiv)
                home_index = np.where(new_indiv == home)[0][0]
                new_indiv[0], new_indiv[home_index] = new_indiv[home_index], new_indiv[0]
                self.currentpop.append(new_indiv)
    def crossOver(self, parents, nbChildren):
        xdim = self.numParameters
        children = []
        for i in xrange(nbChildren):
            p1 = choice(parents)
            if (xdim < 2):
                children.append(p1)
            else:
                p2 = choice(parents)
                
                if len(set(p1)) != len(p1) or len(set(p2)) != len(p2):
                    print p1, p2
                    raise ValueError
                
                point = choice(np.arange(1, xdim-1))
                res = np.zeros(xdim)
                res = p1[:]
                for j in xrange(point, xdim-1):
                    try:
                        v1 = p2[j]
                        v2 = p2[j+1]
                        p1_index = np.where(p1 == v1)[0][0]
                        p2_index = np.where(p1 == v2)[0][0]
                        res[p1_index], res[p2_index] = p1[p2_index], p1[p1_index]
                    except:
                        pass
                if len(set(res)) != len(p1):
                    print res
                    raise ValueError
                children.append(res)
        return children
            
    def mutated(self, indiv):
        """ mutate some genes of the given individual """
        res = indiv.copy()
        for i in xrange(1, self.numParameters-1):
            if random() < self.mutationProb:
                res[i], res[i+1] = indiv[i+1], indiv[i]
        return res
```

Now we can use algorithms for solving problem

```python
def optimize(method, problem, steps=20, minimize=True, verbose=True, temperature=1.0):
    solver = method(lambda x: get_tour_length(x.params), problem, minimize=minimize, verbose=verbose, temperature=temperature)
    res = solver.learn(steps)
    return res
```

```python
def view_tour(mp, best_tour, margin=1):
    '''
    view plot of tour
    '''
    verts = [points[i] for i in best_tour]
    codes = [Path.MOVETO] + [Path.LINETO] * (len(verts)-2) + [Path.CLOSEPOLY]
    x, y = mp
    path = Path(verts, codes)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    patch = patches.PathPatch(path, facecolor='orange', lw=2)
    ax.add_patch(patch)
    ax.set_xlim(x.min() - margin, x.max() + margin)
    ax.set_ylim(y.min() - margin, y.max() + margin)
    plt.show()
```

```python
def do_algo(method, steps=100, temperature=1, verbose=True):
    ''' shortcut '''
    res_tour, best_scores = optimize(method, salesman, steps=steps, minimize=True, temperature=temperature, verbose=verbose)
    view_tour(mp, res_tour.params)
```
