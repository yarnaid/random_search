```python
import numpy as np
```

```python
def col_to_int(col_name, df):
    gr = df.groupby(df[col_name])
    values_dict = dict()
    for k in gr.groups:
        values_dict[k] = df.ix[gr.groups[k]].salary.mean()
    col = df[col_name]
    convert = lambda x: values_dict[x]
    return col.apply(convert)
```

```python
def col_to_int_random(col_name, df):
    col = df[col_name]
    values = set(col)
    numbers = np.arange(len(values))
    values_dict = dict(zip(values, numbers))
    convert = lambda x: values_dict[x]
    return col.apply(convert)
```

```python
cols_to_int = set(df.columns.values[1:])
for col in cols_to_int:
    df[col] = col_to_int(col, df)
cols_to_int = set(df.columns.values[:1])
for col in cols_to_int:
    df[col] = col_to_int_random(col, df)
```

```python
values = df.copy().as_matrix()
```

```python
from pybrain.datasets import SequentialDataSet 
x_dim = len(values.T)
out_dim = 1
```

```python
ds = SequentialDataSet(len(values.T), 1)
```

```python
from pybrain.structure import SigmoidLayer
```

```python
from pybrain.optimization import GA, StochasticHillClimber, HillClimber
```

```python
from pybrain.tools.shortcuts import buildNetwork
neurons_number = ds.indim
hidden = SigmoidLayer
indim = ds.indim
outdim = ds.outdim
gen_nn = buildNetwork(indim, neurons_number, outdim, hiddenclass=hidden)
hc_nn = buildNetwork(indim, neurons_number, outdim, hiddenclass=hidden)
an_nn = buildNetwork(indim, neurons_number, outdim, hiddenclass=hidden)
```

```python
for i in xrange(1000):
    x = values[i]
    y = (salary_df.salary.ix[i], )
    ds.addSample(x, y)
```

```python
ga = GA(ds.evaluateModuleMSE, gen_nn, minimize=True)
hc = HillClimber(ds.evaluateModuleMSE, hc_nn, minimize=True)
an = StochasticHillClimber(ds.evaluateModuleMSE, an_nn, minimize=False)
```
