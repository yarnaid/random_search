```python
def col_to_int(col_name, df):
    gr = df.groupby(df[col_name])
    values_dict = dict()
    for k in gr.groups:
        values_dict[k] = df.ix[gr.groups[k]].salary.mean()
    col = df[col_name]
    convert = lambda x: values_dict[x]
    return col.apply(convert)
```

# Genetic

```python
from scipy import spatial
```

```python
tree = spatial.KDTree(df.ix[:, 1:-1].as_matrix())
def get_gen(gen):
    gen_id = tree.query(gen)[1]
    return gen_id
```

```python
def individual():
    'Create a member of the population.'
    return pd.np.random.randint(0, len(df))

def population(count):
    """
    Create a number of individuals (i.e. a population).

    count: the number of individuals in the population
    """
    return [ individual() for x in xrange(count) ]
```

```python
class Fitness(object):
    def __init__(self, init=None):
        self.best_gen = init if init is not None else np.random.randint(0, len(df))
        self.target = df.iloc[self.best_gen][-1]

    def __call__(self, gen):
        if isinstance(gen, int):
            gen_id = gen
        else:
            gen_id = get_gen(gen)
        new_salary = df.iloc[gen_id][-1]
        ratio = float(new_salary)/self.target
        if ratio > 1:
            self.best_gen = gen_id
            self.target = new_salary
        return ratio
```

```python
from operator import add
```

```python
def grade(population, fitness):
    inv_sum = np.sum([1./fitness(p) for p in population])
    return [1./fitness(p)/inv_sum for p in population]
```

```python
import bisect
```

```python
class WeightedRandomGenerator(object):
    def __init__(self, weights):
        self.totals = []
        running_total = 0

        for w in weights:
            running_total += w
            self.totals.append(running_total)

    def next(self):
        rnd = np.random.random() * self.totals[-1]
        return bisect.bisect_right(self.totals, rnd)

    def __call__(self):
        return self.next()
```

```python
pos_limits = []
for col in df.columns[1:-1]:
    values = set(df[col])
    limits = (min(values), max(values))
    pos_limits.append(limits)

def get_new_mutant(individual):
    pos_to_mutate = np.random.randint(0, len(df.columns[1:-1])-1)
    # this mutation is not ideal, because it
    # restricts the range of possible values,
    # but the function is unaware of the min/max
    # values used to create the individuals,
    new_value = pos_limits[pos_to_mutate][0] + np.random.random() * pos_limits[pos_to_mutate][1]
    new_gen = tree.data[individual]
    new_gen[pos_to_mutate] = new_value
    return get_gen(new_gen)

    
def mutate(parents, mutate_p):
    for i, individual in enumerate(parents):
        if mutate_p > np.random.random():
            parents[i] = get_new_mutant(individual)
    return parents
```

```python
tried = set()
def evolve(parents, fitness, random_select=0.05, mutate_p=0.01):
    parents = mutate(parents, mutate_p)
    # crossover parents to create children
    children = []
    counter = 0
    wrg = WeightedRandomGenerator(grade(parents, fitness))
    while len(children) < len(parents):
        male = parents[wrg()]
        female = parents[wrg()]
        if male != female:
            male = tree.data[male]
            female = tree.data[female]
            half = np.random.randint(1, len(male) - 1)
            child = list(male[:half]) + list(female[half:])
            child = get_gen(child)
            while child in children and child in tried:
                child = get_new_mutant(child)
            children.append(child)
            tried.add(child)
        if counter > len(parents) * 2:
            parents = mutate(parents, 0.5)
    return children
```
