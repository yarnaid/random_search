```python
import numpy as np
```

```python
def col_to_int(col_name, df):
    gr = df.groupby(df[col_name])
    values_dict = dict()
    for k in gr.groups:
        values_dict[k] = df.ix[gr.groups[k]].salary.mean()
    col = df[col_name]
    convert = lambda x: values_dict[x]
    return col.apply(convert)
```

```python
def col_to_int_random(col_name, df):
    col = df[col_name]
    values = set(col)
    numbers = np.arange(len(values))
    values_dict = dict(zip(values, numbers))
    convert = lambda x: values_dict[x]
    return col.apply(convert)
```

```python
cols_to_int = set(df.columns.values[0:-1])
for col in cols_to_int:
    df[col] = col_to_int(col, df)
cols_to_int = set(df.columns.values[:1])
for col in cols_to_int:
    df[col] = col_to_int_random(col, df)
```

# Algorithms

```python
from scipy import spatial
```

```python
class get_n_nearest(object):
    '''
    get the nearest points of current x
    '''
    def __init__(self, df):
        self.tree = spatial.KDTree(df.copy().as_matrix())
        self.df = df.copy()
    
    def __call__(self, x, *params):
        '''
        main function of class
        
        x: vector with dim == dim(df)
        params: list with params[0] == number_of_points_need
        '''
        new_move = self.tree.query(x, params[0])[1]
        return new_move
    
```

## Annealing

```python
from optimisation import annealing
reload(annealing)
```

```python
class BestSalary(annealing.Annealer):
    def __init__(self, state, df, nearest_x_num=None):
        self.tree = spatial.KDTree(df.as_matrix())
        self.df = df
        self.nearest_x_num = nearest_x_num if nearest_x_num is not None else 2  # for simplicity
        self.gnf = get_n_nearest(df)
        super(BestSalary, self).__init__(state)
        
    def energy(self):
        id = self.gnf(self.state, self.nearest_x_num)[0]
        self.gnf.tree.data[id] = -self.gnf.tree.data[id]  # remove visited state
        res = -df_orig.iloc[id][-1]  # we minimize value of energy
        if list(self.state) not in self.gnf.tree.data:  # 0 outside our surface
            res = 0
        return res

    def move(self):
        ''' Let's find just simmilar job by given parameters
        '''
        ids = self.gnf(self.state, self.nearest_x_num)
        new_state_id = ids[0]
        self.state = self.df.iloc[new_state_id]
```

```python
start_x = init_function()
bs = BestSalary(start_x, df)
bs.steps = 200
print 'Started for position ', start_x
state, salary = bs.anneal()
print 'Best job is ', features_df.iloc[bs.gnf(state, 1)], '\nwith salary ', -salary
```

## Genetic

```python
my_gnf = get_n_nearest(df)
def fitness(x):
    x_id = my_gnf(x, 1)
    if not isinstance(x_id, int):
        x_id = x_id[0]
#     my_gnf.tree.data[x_id] = -my_gnf.tree.data[x_id]
    return df_orig.iloc[x_id][-1]
```

```python
def get_random_genome():
    new_genome = np.asarray([np.random.random() for x in xrange(len(my_gnf.tree.data.T))]) * np.asarray(up_lims)
    new_id = my_gnf(new_genome, 1)
    return my_gnf.tree.data[new_id]
```

# Let's play with neural networks
Use `pybrain` for networks.

For every algorithm workflow is following:
1. Get initial $\pmb{x}$
2. Train NN with $\pmb{x}$
3. Run one alogorithm (annealing, climbing or genetic) iteration
4. Get new $\pmb{x}$ or set of $\pmb{x}$
5. Go to step 2. until number of iterations is not exceeded
